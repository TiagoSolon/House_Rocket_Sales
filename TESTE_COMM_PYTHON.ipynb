```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/kc_house_data.csv')

df.head()
```

```python
# use o np.where para criar nova coluna entre 2 condicoes de outra coluna

df['renovated_time'] = np.where(df['yr_renovated']!= 0,  df['yr_renovated'] - df['yr_built'], 0 )
```

```python
df[df['renovated_time'] > 0].head(10)
```

```python
# Use o if e for para criar nova coluna a partir da condicao de outra coluna

df['size_house'] = ['big' if bed > 6 else 'small' for bed in df['bedrooms'] ]
```

```python
# Use o np.select para criar nova coluna a partir de 2 ou mais condicoes

conditionlist = [
    (df['price'] <= 2500000),
    (df['price']  > 2500000) & (df['price']  < 5000000),
    (df['price']  > 5000000)]
choicelist = ['low', 'mid', 'high']
df['padrao'] = np.select(conditionlist, choicelist, default='not specified')
```

```python
# use um dos comandos abaixo para contar a frequencia de valores em uma coluna

#df['padrao'].value_counts()

#df.groupby(['padrao']).count() 

df.groupby(['padrao']).size().rank 
```
